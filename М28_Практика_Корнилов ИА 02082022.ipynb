## 2. Подготовка данных (Data preparation)

```python
df.hist(bins=50, figsize = (15,20))
```

```python
df.info()
```

```python
def print_fullycomplete_and_percent(df):
    print('количество полностью заполненных объектов ',len(df.dropna()))
    print('процент полностью заполненных объектов из всей выборки ',round(len(df.dropna()) / len(df) * 100, 2 ))
print_fullycomplete_and_percent(df)
```

```python
missing_values_percent = ( (df.isna().sum() / len(df))* 100 ).sort_values(ascending=False)
missing_values_percent 
```

```python
#Выведим список всех признаков, в которых имеются пропущенные значения
missing_columns = missing_values_percent[missing_values_percent.values > 0].index.tolist()
missing_columns
```

```python
#Обработаю пропуски в фичах year, удалив записи, в которых год выпуска автомобиля не заполнен, и сохраняю обновлённый датафрейм
df = df[df.year.notna()]
df = df[df.long.notna()]
df = df[df.lat.notna()]
```

```python
#Обработаем пропуски в фиче fuel, заполнив незаполненные значения типа топлива значением other, и сохраним изменения. 
df.fuel = df.fuel.fillna('other')
```

```python
#обработаем пропуски в фиче model, заполнив незаполненные значения модели автомобиля значением other, и сохраните изменения
df.model = df.model.fillna('other')
```

```python
#обработаем пропуски, заполнив незаполненные значения статуса автомобиля самым популярным значением, и сохраните изменения
df.title_status = df.title_status.fillna((df.title_status.mode())[0])
```

```python
def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25 - 1.5*iqr, q75 + 1.5 * iqr)
    return boundaries
```

```python
#Так как выбросы в годе выпуска находятся среди минимальных значений, заполните те значения, которые являются выбросами, значением нижней границы. Не забудьте учесть, что год должен быть целочисленным значением, входящим в границы нормальных значений.
boundaries = calculate_outliers(df.year)
is_outlier = (df.year < boundaries[0]) | (df.year > boundaries[1])
df.loc[is_outlier, 'year'] = int(boundaries[1])
```

```python
print_fullycomplete_and_percent(df)
```

## 3. Feature engineering

```python
ohe = OneHotEncoder(sparse=False)
```

```python
df['short_model'] = df.apply(lambda x: x.model.split(' ')[0], axis=1)
```

```python
ohe.fit(df[['short_model', 'transmission', 'region', 'manufacturer', 'state', 'title_status']])
```

```python
ohe_data = ohe.transform(df[['short_model', 'transmission', 'region', 'manufacturer', 'state', 'title_status']])
```

```python
df[ohe.get_feature_names()] = ohe_data
```

```python
data = df[['lat', 'long', 'year']]
```

```python
std_scaler = StandardScaler()
```

```python
std_scaler.fit(data[['lat', 'long', 'year']])
```

```python
std_scaled = std_scaler.transform(data[['lat', 'long', 'year']])
```

```python
df[['scalled_lat', 'scalled_long', 'scalled_year']] = std_scaled
```

## 4. Modelling

```python
list = df.columns
```

```python
print(list)
```

```python
x = df [['scalled_year','scalled_lat', 'scalled_long']]
y = df['price_category']
```

```python
logreg = LogisticRegression(random_state = 42)
tree = RandomForestClassifier(random_state = 42)
mlp_tuned = MLPClassifier(random_state=42, max_iter=500, hidden_layer_sizes=(100,20), activation = 'tanh')
```

```python
for model in [logreg, tree, mlp_tuned]:
    cv_score = cross_val_score(model, x, y, cv=5)
    print(cv_score.mean(), cv_score.std())
```

## Results

```python
#наилучший результат дал MLPClassifier многослойный персептрон
```

```python
mlp_tuned.fit(x, y)

pred_mlp2 = mlp_tuned.predict(x)

print(accuracy_score(y, pred_mlp2))
```
